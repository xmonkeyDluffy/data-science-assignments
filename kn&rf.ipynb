```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv("heart.csv")
data
```

```python
X=data.iloc[:,:-1]
X
```

```python
Y=data.iloc[:,-1]
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,Y_train)
```

```python
y_predict_knn=knn.predict(X_test)
```

```python
accuracy_score(y_predict_knn,Y_test)
```

```python
errors=[]
for i in range(5,21):
  knn1=KNeighborsClassifier(i)
  knn1.fit(X_train,Y_train)
  knn1_predict=knn1.predict(X_test)
  errors.append(accuracy_score(Y_test,knn1_predict))
```

```python
max_ac=max(errors)
print(errors.index(max_ac)+5)
```

```python
print(*errors)
```

```python
print(max(errors))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest=RandomForestClassifier()
```

```python
random_forest.fit(X_train,Y_train)
```

```python
y_predict_random_forest=random_forest.predict(X_test)
```

```python
accuracy_score(Y_test,y_predict_random_forest)
```
